```python
import pandas as pd

# Load data (update path if needed)
data = pd.read_csv('accepted_2007_to_2018Q4.csv')

# Shape
print('Shape:', data.shape)

# Numerical summary for key features
key_num = ['loan_amnt', 'int_rate', 'annual_inc', 'dti', 'fico_range_low']
print(data[key_num].describe())

# Categorical counts
print('Grade counts:\n', data['grade'].value_counts())
print('Loan status counts:\n', data['loan_status'].value_counts())

# Missing values %
missing = data.isnull().mean() * 100
print('Missing %:\n', missing[missing > 0].sort_values(ascending=False))
```
